```python
data["Age"].fillna(data["Age"].median(),inplace=True)
data["Income"].fillna(data["Income"].median())
data["Spending_Score"].fillna(data["Spending_Score"].median())
data["City"].fillna(data["City"].mode()[0])
print(data)
```

```python
data.drop_duplicates(inplace=True)
```

```python
data["Purchased"]=data["Purchased"].astype("category")
print(data)
```

```python
plt.scatter(data["Income"],data["Purchased"])
```

```python
Q1=data['Income'].quantile(0.25)
Q2=data['Income'].quantile(0.75)
IQR=Q2-Q1
print(Q1,Q2,IQR)
print(data["Income"])
```

```python
data=pd.get_dummies(data,columns=['City'], drop_first=True, dtype=int)
print(data)
data.head()
```
